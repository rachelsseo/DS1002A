```python
#RGB HW4 START

import numpy as np
from PIL import Image
import requests
```

```python
#img 1

url1 = 'https://images.nationalgeographic.org/image/upload/t_edhub_resource_key_image/v1638886205/EducationHub/photos/rainbow-river.jpg'
image = np.asarray(Image.open(requests.get(url1, stream=True).raw))
```

```python
image
```

```python
#array for image
print(image)
```

```python
#red filtered photo
def red_only(img):
  red_image = image.copy()
  for i in red_image:
    for j in i:
      j[1] = 0
      j[2] = 0
  return red_image

#green filtered photo
def green_only(img):
  green_image = image.copy()
  for i in green_image:
    for j in i:
      j[0] = 0
      j[2] = 0
  return green_image

#blue filtered photo
def blue_only(img):
  blue_image = image.copy()
  for i in blue_image:
    for j in i:
      j[0] = 0
      j[1] = 0
  return blue_image

#red green filtered photo
def red_green_only(img):
  red_green_image = image.copy()
  for i in red_green_image:
    for j in i:
      j[2] = 0
  return red_green_image

#red blue filtered photo
def red_blue_only(img):
  red_blue_only = image.copy()
  for i in red_blue_only:
    for j in i:
      j[1] = 0
  return red_blue_only

#blue green filtered photo
def blue_green(img):
  blue_green_only = image.copy()
  for i in blue_green_only:
    for j in i:
      j[0] = 0
  return blue_green_only

def output(img):
  red_img = red_only(img)
  green_img = green_only(img)
  blue_img = blue_only(img)
  red_green_img = red_green_only(img)
  red_blue_img = red_blue_only(img)
  green_blue_img = blue_green(img)

  print("Original Image", "\n", img)
  print("Red Image:", "\n", red_img)
  print("Green Image:", "\n", green_img)
  print("Blue Image:", "\n", blue_img)
  print("Red Green Image:", "\n", red_green_img)
  print("Red Blue Image:", "\n", red_blue_img)
  print("Green Blue Image:", "\n", green_blue_img)

  #Logical Check
  print("Logical Check:")

  sum_img = (red_img + blue_img + green_img)
  if(np.array_equal(sum_img, img)):
    print("Arrays (images) are equal")
  else:
    print("Arrays (images) are not equal")
```

```python
#image 1 output:
output(image)
```
